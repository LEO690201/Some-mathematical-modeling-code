```python
import matplotlib.pyplot as plt  # 导入matplotlib库用于绘图
import networkx as nx   # 导入networkx库，适用于复杂网络分析

# 无权重无向图
s1 = [1, 2, 3, 4]  # 定义边的起点列表
t1 = [2, 3, 1, 1]  # 定义边的终点列表
G1 = nx.Graph()  # 创建一个空的无向图
G1.add_edges_from(zip(s1, t1))  # 添加边到图中
plt.figure()  # 创建一个新的图形
nx.draw(G1, with_labels=True)  # 绘制图，节点带标签
plt.gca().set_xticks([])  # 隐藏x轴刻度
plt.gca().set_yticks([])  # 隐藏y轴刻度
plt.show()  # 显示图形

# 有权重无向图
s = [1, 2, 3, 4]  # 定义边的起点列表
t = [2, 3, 1, 1]  # 定义边的终点列表
w = [3, 8, 9, 2]  # 定义边的权重列表
G = nx.Graph()  # 创建一个空的无向图
G.add_weighted_edges_from(zip(s, t, w))  # 添加带权重的边到图中
plt.figure()  # 创建一个新的图形
pos = nx.spring_layout(G)  # 使用spring布局算法计算节点位置
nx.draw(G, pos, with_labels=True, node_size=500, edge_color='b', width=2)  # 绘制图，节点带标签，设置节点大小和边的颜色、宽度
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}  # 提取每条边的权重作为标签
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)  # 在边上绘制权重标签
plt.gca().set_xticks([])  # 隐藏x轴刻度
plt.gca().set_yticks([])  # 隐藏y轴刻度
plt.show()  # 显示图形

# 无权重有向图
s = [1, 2, 3, 4, 1]  # 定义边的起点列表
t = [2, 3, 1, 1, 4]  # 定义边的终点列表
G = nx.DiGraph()  # 创建一个空的有向图
G.add_edges_from(zip(s, t))  # 添加边到图中
plt.figure()  # 创建一个新的图形
nx.draw(G, with_labels=True)  # 绘制图，节点带标签
plt.gca().set_xticks([])  # 隐藏x轴刻度
plt.gca().set_yticks([])  # 隐藏y轴刻度
plt.show()  # 显示图形

# 有权重有向图
s = [1, 2, 3, 4]  # 定义边的起点列表
t = [2, 3, 1, 1]  # 定义边的终点列表
w = [3, 8, 9, 2]  # 定义边的权重列表
G = nx.DiGraph()  # 创建一个空的有向图
G.add_weighted_edges_from(zip(s, t, w))  # 添加带权重的边到图中
plt.figure()  # 创建一个新的图形
pos = nx.spring_layout(G)  # 使用spring布局算法计算节点位置
nx.draw(G, pos, with_labels=True, node_size=500, edge_color='b', width=2)  # 绘制图，节点带标签，设置节点大小和边的颜色、宽度
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}  # 提取每条边的权重作为标签
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)  # 在边上绘制权重标签
plt.gca().set_xticks([])  # 隐藏x轴刻度
plt.gca().set_yticks([])  # 隐藏y轴刻度
plt.show()  # 显示图形
```
