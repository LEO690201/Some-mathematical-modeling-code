```python
import matplotlib.pyplot as plt  # matplotlib
import networkx as nx   # networkx

# 
s1 = [1, 2, 3, 4]  # 
t1 = [2, 3, 1, 1]  # 
G1 = nx.Graph()  # 
G1.add_edges_from(zip(s1, t1))  # 
plt.figure()  # 
nx.draw(G1, with_labels=True)  # 
plt.gca().set_xticks([])  # x
plt.gca().set_yticks([])  # y
plt.show()  # 

# 
s = [1, 2, 3, 4]  # 
t = [2, 3, 1, 1]  # 
w = [3, 8, 9, 2]  # 
G = nx.Graph()  # 
G.add_weighted_edges_from(zip(s, t, w))  # 
plt.figure()  # 
pos = nx.spring_layout(G)  # spring
nx.draw(G, pos, with_labels=True, node_size=500, edge_color='b', width=2)  # 
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}  # 
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)  # 
plt.gca().set_xticks([])  # x
plt.gca().set_yticks([])  # y
plt.show()  # 

# 
s = [1, 2, 3, 4, 1]  # 
t = [2, 3, 1, 1, 4]  # 
G = nx.DiGraph()  # 
G.add_edges_from(zip(s, t))  # 
plt.figure()  # 
nx.draw(G, with_labels=True)  # 
plt.gca().set_xticks([])  # x
plt.gca().set_yticks([])  # y
plt.show()  # 

# 
s = [1, 2, 3, 4]  # 
t = [2, 3, 1, 1]  # 
w = [3, 8, 9, 2]  # 
G = nx.DiGraph()  # 
G.add_weighted_edges_from(zip(s, t, w))  # 
plt.figure()  # 
pos = nx.spring_layout(G)  # spring
nx.draw(G, pos, with_labels=True, node_size=500, edge_color='b', width=2)  # 
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}  # 
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)  # 
plt.gca().set_xticks([])  # x
plt.gca().set_yticks([])  # y
plt.show()  # 
```
