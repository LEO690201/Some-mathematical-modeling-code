```python
import numpy as np
import matplotlib.pyplot as plt

# ,10000,1,(1,1)
p=10000 #
r=1 #
x0=1 #x
y0=1 #y
n=0  #:0

# 
plt.figure() #
plt.title('Monte Carlo Method') #
plt.xlabel('x') #x
plt.ylabel('y') #y

# 
for i in range(p):
    # np.random.rand()(0,1)
    px=np.random.rand()*2  #
    py=np.random.rand()*2  #

    # 
    if(px-x0)**2+(py-y0)**2<=r**2:
        plt.plot(px,py,'.',color='b')
        n+=1
    else:
        plt.plot(px,py,'.',color='r')

plt.axis('equal') #
plt.show() #


# 
s=(n/p)*4
pi_estimate=s
print(':',s)
print(':',pi_estimate)

```
