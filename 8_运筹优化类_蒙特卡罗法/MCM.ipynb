```python
import numpy as np
import matplotlib.pyplot as plt

# 参数初始化,投放10000个点,圆半径为1,圆心坐标为(1,1)
p=10000 #总共投放的点数
r=1 #圆半径
x0=1 #圆心x坐标
y0=1 #圆心y坐标
n=0  #初始化:开始时有0个点落在圆内

# 设置绘图窗口
plt.figure() #创建绘图窗口
plt.title('Monte Carlo Method') #设置标题
plt.xlabel('x') #设置x轴标签
plt.ylabel('y') #设置y轴标签

# 保持图形窗口多次绘图
for i in range(p):
    # np.random.rand()生成(0,1)之间的随机数
    px=np.random.rand()*2  #随机生成该点的横坐标
    py=np.random.rand()*2  #随机生成该点的纵坐标

    # 判断点是否在圆内
    if(px-x0)**2+(py-y0)**2<=r**2:
        plt.plot(px,py,'.',color='b')
        n+=1
    else:
        plt.plot(px,py,'.',color='r')

plt.axis('equal') #设置坐标轴比例一致
plt.show() #显示绘图结果


# 计算圆内点的比例
s=(n/p)*4
pi_estimate=s
print('圆内点的比例:',s)
print('估计π值:',pi_estimate)

```

```python
import numpy as np

# 预备知识
# np.random.randint(low, high, size)在指定区间内生成随机整数
# np.random.randint(1,4)生成1到3之间的随机整数

n=10000
a=0
b=0
for i in range(n):
    x=np.random.randint(1,4)
    y=np.random.randint(1,4)
    if x==y:
        a+=1
    else:
        b+=1

print('蒙特卡罗法不改变注意的概率:',a/n)
print('蒙特卡罗法改变注意的概率:',b/n)

# 如果考虑失败情况:
n=10000
a=0
b=0
c=0
for i in range(n):
    x=np.random.randint(1,4)
    y=np.random.randint(1,4)
    change=np.random.randint(0,2)
    if x==y:
        if change==0:
            a+=1
        else:
            c+=1
    else:
        if change==0:
            c+=1
        else:
            b+=1

print('蒙特卡罗法不改变注意的概率:',a/n)
print('蒙特卡罗法改变注意的概率:',b/n)
print('蒙特卡罗法改变注意且失败的概率:',c/n)





```
